# MODEL BUILDING

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_validate
from sklearn.metrics import roc_curve,auc
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

### Note: Sampling the data to 5% of actual dataset as it is taking huge time to run the models

```python
data_sampled = data.sample(frac =.05)


# the data is still imbalanced with majority class 0 and minority class 1 values
(data_sampled.target==0).sum()
```

```python
scoring = {'Accuracy' : make_scorer(accuracy_score), 
           'Precision' : make_scorer(precision_score),
           'Recall' : make_scorer(recall_score), 
           'F1_Score' : make_scorer(f1_score)}
           

#kfold = KFold(n_splits=10, random_state=0, shuffle=True)
kfold = StratifiedKFold(n_splits=10)

#data.reset_index(inplace = True)
df = data_sampled.copy()
Y = df["target"]
df = df.drop('target', 1)
df = df.drop('id', 1)
X = df


```

###### (i) No balancing technique

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

model1 = LogisticRegression(solver='liblinear')
predictions = cross_val_predict(model1, X, Y, cv=10)
results = cross_validate(estimator=model1, X=X, y=Y, cv=kfold, scoring=scoring)
print("Accuracy: ",np.mean(results['test_Accuracy']))
print("Precision: ",np.mean(results['test_Precision']))
print("Recall: ",np.mean(results['test_Recall']))
print("F1 score: ",np.mean(results['test_F1_Score']))
auc = cross_val_score(model1, X, Y, scoring="roc_auc", cv = kfold).mean()
print('auc=', auc)
```

```python
# Naive bayes
from sklearn.naive_bayes import GaussianNB

model2 = GaussianNB()
results = cross_validate(estimator=model2, X=X, y=Y, cv=kfold, scoring=scoring)
predictions = cross_val_predict(model2, X, Y, cv=10)
print("Accuracy: ",np.mean(results['test_Accuracy']))
print("Precision: ",np.mean(results['test_Precision']))
print("Recall: ",np.mean(results['test_Recall']))
print("F1 score: ",np.mean(results['test_F1_Score']))
auc = cross_val_score(model2, X, Y, scoring="roc_auc", cv = kfold).mean()
print('auc=', auc)
```

```python
from sklearn.ensemble import RandomForestClassifier

model3 = RandomForestClassifier()
results = cross_validate(estimator=model3, X=X, y=Y, cv=kfold, scoring=scoring)
predictions = cross_val_predict(model3, X, Y, cv=10)
print("Accuracy: ",np.mean(results['test_Accuracy']))
print("Precision: ",np.mean(results['test_Precision']))
print("Recall: ",np.mean(results['test_Recall']))
print("F1 score: ",np.mean(results['test_F1_Score']))
auc = cross_val_score(model3, X, Y, scoring="roc_auc", cv = kfold).mean()
print('auc=', auc)
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
model4 = BalancedRandomForestClassifier()
results = cross_validate(estimator=model4, X=X, y=Y, cv=kfold, scoring=scoring)
predictions = cross_val_predict(model4, X, Y, cv=10)
print("Accuracy: ",np.mean(results['test_Accuracy']))
print("Precision: ",np.mean(results['test_Precision']))
print("Recall: ",np.mean(results['test_Recall']))
print("F1 score: ",np.mean(results['test_F1_Score']))
auc = cross_val_score(model4, X, Y, scoring="roc_auc", cv = kfold).mean()
print('auc=', auc)
```

```python
from imblearn.ensemble import RUSBoostClassifier
model5 = RUSBoostClassifier()
results = cross_validate(estimator=model5, X=X, y=Y, cv=kfold, scoring=scoring)
predictions = cross_val_predict(model5, X, Y, cv=10)
print("Accuracy: ",np.mean(results['test_Accuracy']))
print("Precision: ",np.mean(results['test_Precision']))
print("Recall: ",np.mean(results['test_Recall']))
print("F1 score: ",np.mean(results['test_F1_Score']))
auc = cross_val_score(model5, X, Y, scoring="roc_auc", cv = kfold).mean()
print('auc=', auc)
```

###### In case of no rebalancing technique used we get high precision values and low recall values as model is biased due to majority class 0 values and  minority  class 1 values. But for BalancedRandomForestClassifier and RUSBoostClassifier the recall values are good and overall resulats are poor because these 2 models use rebalancing techniques internally to balance the data, hence they give good values for minority class (for our case minority is class 1 and hence better recall values)

###### Results are better with oversampling when compared to undersampling as undersampling eliminates most of the data from majority class (class 0) and duplicates data of minority class (class 1). Due to this there can be chances that important data might be lost.  In cases where minority class instances are very low using oversampling is preferred.

###### Logistic regression gives good accuracy values as it predicts majority class (class 0) accurately but poor precision, recall, F1 score when compared to other classifiers when no rebalancing technique is used, as it cannot handle imbalanced data. BalancedRandomForestClassifier and RUSBoostClassifier give good performance as they internally balance the data and boost the performance. When undersampling technique is used logistic regression performs almost similar (with slightly lower values) compared to other classifiers. When oversampling technique is used other models perform better and give higher accuracy, precision, recall values when compared to logistic regression model.



###### RandomForestClassifier and BalancedRandomForestClassifier perform differently when no data rebalancing technique is used i.e. BalancedRandomForestClassifier performs better as it internally uses rebalancing technique to handle imbalanced data. After using different rebalancing techniques both BalancedRandomForestClassifier and RandomForestClassifier give same performances as now both imbalanced data is handled and given to classifiers.
